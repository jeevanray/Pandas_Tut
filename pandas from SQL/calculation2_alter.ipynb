```python
def get_interpolation(start_year, end_year, start_val, end_val):
    if start_year <= end_year:
        years = np.arange(start_year, end_year + 1)
        if start_val != end_val:
            values = np.linspace(start_val, end_val, len(years))
        else:
            values = np.full(len(years), start_val)
        return pd.DataFrame({'interp_year': years, 'interp_value': values})
    else:
        return pd.DataFrame()
```

```python
interp_country_indirect_probability_set = pd.concat([
    get_interpolation(
        row['Year'], 
        row['lead_year'] if row['lead_year'] is not None else row['Year'],
        row['indirect_probability'], 
        row['lead_indirect_probability'] if row['lead_indirect_probability'] is not None else row['indirect_probability']
    ).assign(entity_id=row['entity_id'], scenario_code=row['Scenario'], country=row['Country'])
    for _, row in z.iterrows()
])
interp_country_indirect_probability_set
```

```python
df['expected_indirect_loss_amount_series_undiscounted'] = np.where(
    (df['revenue_avg_usd'] < 0) | (df['gppe_avg_usd'] < 0) | (df['avg_country_revenue_usd'] < 0), 
    np.nan, 
    (df['interp_FMI'] - df['interp_country_productive_capacity_loss'] * df['interp_FMI']) * df['avg_country_revenue_usd']
)

df['expected_indirect_loss_amount_series_discounted'] = df['expected_indirect_loss_amount_series_undiscounted'] / (1.07**(df['phys_year'] - 2021))

df['expected_revenue_loss_amount_series_undiscounted'] = np.where(
    (df['revenue_avg_usd'] < 0) | (df['gppe_avg_usd'] < 0) | (df['avg_country_revenue_usd'] < 0), 
    np.nan, 
    df['interp_country_productive_capacity_loss'] * df['avg_country_revenue_usd']
)

df['expected_revenue_loss_amount_series_discounted'] = df['expected_revenue_loss_amount_series_undiscounted'] / (1.07**(df['phys_year'] - 2021))

df['expected_asset_damage_loss_amount_series_undiscounted'] = np.where(
    (df['revenue_avg_usd'] < 0) | (df['gppe_avg_usd'] < 0), 
    np.nan, 
    df['interp_country_asset_damage_risk'] * df['gppe_avg_usd'] * df['avg_country_asset_proportion']
)

df['expected_asset_damage_loss_amount_series_discounted'] = df['expected_asset_damage_loss_amount_series_undiscounted'] / (1.07**(df['phys_year'] - 2021))
```

```python
final_df = df[[
    'entity_id', 'scenario_code', 'phys_year', 'country',
    'expected_indirect_loss_amount_series_undiscounted',
    'expected_indirect_loss_amount_series_discounted',
    'expected_revenue_loss_amount_series_undiscounted',
    'expected_revenue_loss_amount_series_discounted',
    'expected_asset_damage_loss_amount_series_undiscounted',
    'expected_asset_damage_loss_amount_series_discounted',
    'interp_country_productive_capacity_loss'
]]
final_df
```

```python
aggregated_df = final_df.groupby(
    ['entity_id', 'scenario_code', 'phys_year']
).agg(
    expected_revenue_loss_amount_series_undiscounted=pd.NamedAgg(column='expected_revenue_loss_amount_series_undiscounted', aggfunc='sum'),
    expected_asset_damage_loss_amount_series_undiscounted=pd.NamedAgg(column='expected_asset_damage_loss_amount_series_undiscounted', aggfunc='sum'),
    expected_indirect_loss_amount_series_undiscounted=pd.NamedAgg(column='expected_indirect_loss_amount_series_undiscounted', aggfunc='sum'),
    expected_revenue_loss_amount_series_discounted=pd.NamedAgg(column='expected_revenue_loss_amount_series_discounted', aggfunc='sum'),
    expected_asset_damage_loss_amount_series_discounted=pd.NamedAgg(column='expected_asset_damage_loss_amount_series_discounted', aggfunc='sum'),
    expected_indirect_loss_amount_series_discounted=pd.NamedAgg(column='expected_indirect_loss_amount_series_discounted', aggfunc='sum')
).reset_index()
aggregated_df
```

```python
aggregated_df['expected_direct_damage_loss_amount_series_undiscounted'] = (
    aggregated_df['expected_revenue_loss_amount_series_undiscounted'] + 
    aggregated_df['expected_asset_damage_loss_amount_series_undiscounted']
)

aggregated_df['expected_direct_damage_loss_amount_series_discounted'] = (
    aggregated_df['expected_revenue_loss_amount_series_discounted'] + 
    aggregated_df['expected_asset_damage_loss_amount_series_discounted']
)

aggregated_df['expected_total_loss_amount_series_undiscounted'] = (
    aggregated_df['expected_direct_damage_loss_amount_series_undiscounted'] + 
    aggregated_df['expected_indirect_loss_amount_series_undiscounted']
)

aggregated_df['expected_total_loss_amount_series_discounted'] = (
    aggregated_df['expected_direct_damage_loss_amount_series_discounted'] + 
    aggregated_df['expected_indirect_loss_amount_series_discounted'] )

aggregated_df = aggregated_df.sort_values(by=['entity_id', 'scenario_code', 'phys_year'])
```

```python
aggregated_df = aggregated_df.rename(columns = {"scenario_code" :"rcp_scenario", "phys_year" : "physical_impact_year"
})
aggregated_df['physical_impact_year'] = aggregated_df['physical_impact_year'].astype(int)
```
