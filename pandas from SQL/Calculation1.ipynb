```python
def categorize_level(level, tier_conditions,tier_type):
    if pd.isna(level):
        return None
    tier_country = tier_conditions[tier_conditions['tier_type'] == tier_type].iloc[:,:-1].reset_index(drop=True)
    for _, row in tier_country.iterrows():
        if level is None:
            return None
        if row['lowerbound'] < level <= row['upperbound']:
            return row['tier']
    return None
```

```python
# Use numpy.where to conditionally select the country_iso3 based on the maximum values
top_country['top_country_contributor_for_asset_damage_risk'] = np.where(
    top_country['country_contribution_percentage_asset_damage_risk'] == top_country['max_asset_damage_risk'],
    top_country['country_iso3'], 
    np.nan
)

top_country['top_country_contributor_for_high_risk_assets'] = np.where(
    top_country['country_contribution_percentage_high_risk_assets'] == top_country['max_high_risk_assets'],
    top_country['country_iso3'], 
    np.nan
)

top_country['top_country_contributor_for_productive_capacity_loss'] = np.where(
    top_country['country_contribution_percentage_productive_capacity_loss'] == top_country['max_productive_capacity_loss'],
    top_country['country_iso3'], 
    np.nan
)

top_country['top_country_contributor_for_regional_risk'] = np.where(
    top_country['country_contribution_percentage_regional_infrastructure_risk'] == top_country['max_regional_infrastructure_risk'],
    top_country['country_iso3'], 
    np.nan
)

top_country['top_country_contributor_for_local_critical_infrastructure'] = np.where(
    top_country['percentage_local_critical_infrastructure_risk'].fillna(-1) == top_country['max_local_critical_infrastructure_risk'],
    top_country['country_iso3'], 
    np.nan
)
```
